```python
from transformers import pipeline
```

#Step 1: Text Generation Step

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

prompt = "The future of Artificial Intelligence is"

for name, model in models.items():
    print(f"\n{name}:")
    try:
        generator = pipeline("text-generation", model=model)
        result = generator(prompt, max_length=30, num_return_sequences=1)
        print(result[0]['generated_text'])
    except Exception as e:
        print("Failed:", e)
```

Fill Mask

```python
sentence = "The goal of Generative AI is to <mask> new content."

for name, model in models.items():
    print(f"\n{name}:")
    try:
        masker = pipeline("fill-mask", model=model)
        for pred in masker(sentence)[:3]:
            print(pred['token_str'])
    except Exception as e:
        print("Failed:", e)
```

Question Answering

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for name, model in models.items():
    print(f"\n{name}:")
    try:
        qa = pipeline("question-answering", model=model)
        answer = qa(question=question, context=context)
        print(answer["answer"])
    except Exception as e:
        print("Failed:", e)
```
